```python
from urllib.request import urlopen
```

```python
from bs4 import BeautifulSoup as soup
```

```python
url_link = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area"
```

```python
data = urlopen(url_link)
```

```python
data_html = data.read()
```

```python
data.close()
```

```python
parsed_html = soup(data_html,"html.parser")
```

```python
table = parsed_html.find_all("table",{"class":"wikitable sortable sticky-header col2left"})
```

```python
table = table[0]
```

```python
full_table = table.find_all("tr",{})
```

```python
headings = []
for i in full_table[0].find_all("th",{}):
    headings.append(i.text)

print(headings)
```

```python
rows = []
for i in full_table:
    each_row = []
    row = i.find_all("td",{})
    for j in row:
        each_row.append(j.text)
    rows.append(each_row)

print(rows)
```

```python
for i in range(len(headings)):
    headings[i] = headings[i].replace("\n","")
print(headings)
```

```python
for i in range(len(rows)):
    for j in range(len(rows[i])):
        rows[i][j] = rows[i][j].replace(",","")

print(rows)
```

```python
with open("C:\\Users\\Vaibhav Jamdagni\\Downloads\\Data Science\\File Handle\\csv_scraped_hw.csv","w") as file:
    header = ",".join(headings) + "\n"
    file.write(header)

    for i in rows:
        row = ",".join(i)
        file.write(row)
```
